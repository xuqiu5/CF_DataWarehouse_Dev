```python
# import packages
import csv
import pandas as pd
import re
import datetime
import io
from io import StringIO
import string
import numpy as np
```

```python
#Please edit pathway before use
import_path="~/desktop/Data_Cleaning/Demographic_Info/"
export_path="~/desktop/Data_Cleaning/Cleaned/"
#Dictionary
y_n = {
    2: 'Yes',
    1: 'No'
}

bev_frequency = {
    0: 'Never or less than 1 time per week', 
    1: '1 time per week', 
    2: '2-3 times per week',
    3: '4-6 times per week',
    4: '1 time per day',
    5: '2+ times per day',
    6: '3+ times per day',
}

bev_amount ={
    0: 'Less than 6 fl oz (3/4 cup)', 
    1: '8 fl oz (1 cup)', 
    2: '12 fl oz (1 1/2 cups)',
    3: '16 fl oz (2 cups)',
    4: 'More than 20 fl oz (2 1/2 cups)',
}

location ={
    1: 'Augusta University',
    2: 'Emory University',
}

patient ={
    4:'CFBR-AU-0008',
    5:'CFBR-AU-0002',
    6:'CFBR-AU-0004',
    7:'CFBR-AU-0005',
    8:'CFBR-AU-0003',
    9:'CFBR-AU-0006',
    10:'CFBR-AU-0007',
    11:'CFBR-EM-0100'
}
```

```python
#Read in table
sample = pd.read_csv (import_path+'BEVQ15_DiabetesGINutritionA_DATA_2021-09-09_1327.csv')
# clean empty data
sample = sample.replace(np.nan, '', regex = True)
#change column names
sample.rename(columns={
    'record_id':'patient_id',
    'survey_date_bev': 'survey_date_bevq15',
}, inplace=True)
#drop unused columns
sample=sample.drop(['beverage_questionnaire_bevq15_timestamp','redcap_survey_identifier'], axis=1)
```

```python
# change option number data into text
sample = sample.replace({
'patient_id': patient,
'location_bev': location,
'water_often_bev': bev_frequency, 
'water_amount_bev': bev_amount,
'juice_often_bev': bev_frequency, 
'juice_amount_bev': bev_amount,
'sweet_juice_often_bev': bev_frequency, 
'sweet_juice_amount_bev': bev_amount,
'whole_milk_often_bev': bev_frequency, 
'whole_milk_amt_bev': bev_amount,
'red_fat_milk_often_bev': bev_frequency, 
'red_fat_milk_amt_bev': bev_amount,
'fat_free_milk_often_bev': bev_frequency, 
'fat_free_milk_amt_bev': bev_amount,
'soft_drinks_often_bev': bev_frequency, 
'soft_drinks_amount_bev': bev_amount,
'diet_drink_often_bev': bev_frequency, 
'diet_drinks_amount_bev': bev_amount,
'sweet_tea_often_bev': bev_frequency, 
'sweet_tea_amount_bev': bev_amount,
'tea_w_cream_often_bev': bev_frequency, 
'tea_w_cream_amt_bev': bev_amount,
'tea_black_often_bev': bev_frequency, 
'tea_black_amt_bev': bev_amount,
'beer_often_bev': bev_frequency, 
'beer_amount_bev': bev_amount,
'liqour_often_bev': bev_frequency, 
'liqour_amt_bev': bev_amount,
'wine_often_bev': bev_frequency, 
'wine_amount_bev': bev_amount,
'energy_often_bev': bev_frequency, 
'energy_amout_bev': bev_amount,
'other_often_bev': bev_frequency, 
'other_amt_bev': bev_amount,
'beverage_questionnaire_bevq15_complete': y_n,
})
```

```python
#switch date format to mm/dd/yyyy
#date information stored in variable year, month day
date_new=[]
sample["survey_date_bevq15"]

for i in sample["survey_date_bevq15"]:
    if i=='':
        date_new.append('NA')
    else:
        year=re.findall("^[0-9]{4}",i)
        month=re.findall("-[0-9]{2}-",i)[0].strip('-')
        day=re.findall("[0-9]{2}$",i)
        date_new.append(month+'/'+day[0]+'/'+year[0])
    
sample=sample.drop(columns=['survey_date_bevq15'])
sample['survey_date_bevq15']=date_new

#find today's date
today = datetime.datetime.today().strftime('%m-%d-%Y')
```

```python
#export as csv
sample.to_csv (export_path+'beverage_question_BEVQ15'+today+'.csv', index = False, header=True)
```
