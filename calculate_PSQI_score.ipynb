```python
#score 2: sleep latency: num_mins_ls=

num_mins_ls=[]
def convert_mins_to_score(mins):
    if mins=='':
        score_mins=mins
    else:
        if int(mins) <= 15:
            score_mins=0
        elif int(mins) <= 30:
            score_mins=1
        elif int(mins) <= 60:
            score_mins=2
        elif int(mins) > 60:
            score_mins=3
    return(score_mins)

for mins in samples["number_of_mins_psqi"]:
    score=convert_mins_to_score(mins)
    num_mins_ls.append(score)
num_mins_ls

list1 = [1, 2, 3]
list2 = [4, 5, 6]

zipped_lists = zip(num_mins_ls, list2)

sum = [x + y for (x, y) in zipped_lists]
Create a list with the sum of each pair


print(sum)
```
